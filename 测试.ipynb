```python
def maxSlidingWindow(nums, k): #单调递减队列
    n = len(nums)
    q = collections.deque()
    for i in range(k):
        while q and q[-1] < nums[i]:
            q.pop()
        q.append(nums[i])
    ans = [q[0]] #keep 前三个里面最大的
    for i in range(k, len(nums)):
        if q[0] == nums[i-k]:
            q.popleft()
        while q and q[-1] < nums[i]:
            q.pop()
        q.append(nums[i])
        ans.append(q[0])
    return ans
```

```python
def lengthofLongestSubstringKDistince(s, k):
    n = len(s)
    if k == 0 or n == 0: return 0
    
    left, right = 0, 0
    hashmap = collections.defaultdict()
    
    max_len = 1
    
    while right < n:
        hashmap[s[right]] = right #{L:0, O:1} 直接覆盖right的位置
        right += 1
        
        if len(hashmap) == k + 1: #如果超过k个不同字符
            del_idx = min(hashmap.values())
            del hashmap[s[del_idx]] # 删除最前面的字符（value最小的）
            left = del_idx + 1 # left指针右移
        
        if n - left < max_len: break #优化
        
        max_len = max(max_len, right-left)
    return max_len
```

```python
def isValid(self, s: str) -> bool:
    dic = {'{': '}',  '[': ']', '(': ')', '?': '?'}
    stack = ['?']
    for c in s:
        if c in dic: stack.append(c)
        elif dic[stack.pop()] != c: return False 
    return len(stack) == 1

def longestValid(s):
    if not s: return 0
    stack = [-1]
    res = 0
    for i in range(len(s)):
        if s[i] == '(':
            stack.append(i) #下标入栈
        else:
            stack.pop()
            if not stack: stack.append(i) 
            else: res = max(res, i - stack[-1]) 
            #符合消去成对括号时，拿当前下标减去栈顶下标
    return res
```

```python
def generateparenthesis(n):
    if n <= 0: return []
    res = []
    def dfs(cur_str, left, right):
        if left == n and right == n:
            res.append(cur_str)
            return
        if left < right: #left0,right1 ")" 不符合题意，剪枝
            return 
        if left < n:
            dfs(cur_str+'(', left+1, right)
        if right < n:
            dfs(cur_str+'(', left, right+1)
    dfs('',0,0)
    return res
```

leetcode 29 两数相除 二进制搜索 logn * logn
输入: dividend = 10, divisor = 3 输出: 3 不用乘法、除法和 mod 运算符。

```python
def divide(dividend, divisor):
    min_int, max_int = -2**31, 2**31-1
    flag = 1
    if dividend < 0: flag, dividend = -flag, -dividend
    if divisor < 0: flag, divisor = -flag, -divisor
    #每次利用加法，将当前的divisor乘以两倍，并同时用multiple记录下乘以了2的多少次方
    res = 0
    while dividend >= divisor:
        cur = divisor
        multiple = 1
        while cur + cur < dividend:
            cur += cur #cur分别乘以2,4,8
            multiple += multiple
        dividend -= cur
        res += multiple
    res = res if flag > 0 else -res
    if res < min_int:
        return min_int
    elif min_int <= res <= max_int:
        return res
    else:
        return max_int
```

```python
def isvalidSudoku(board):
    rows = [{} for i in range(9)]
    cols = [{} for i in range(9)]
    boxes = [{} for i in range(9)]
    for i in range(9):
        for j in range(9):
            num = board[i][j]
            if num != '.':
                num = int(num)
                box_index = (i//3) * 3 + j // 3
                rows[i][num] = rows[i].get(num, 0) + 1
                cols[i][num] = cols[i].get(num, 0) + 1
                boxes[box_index][num] = boxes[box_index].get(num, 0) + 1
                
                if rows[i][num] > 1 or cols[i][num] > 1 or boxes[box_index][num] > 1:
                    return False
    return True
```

```python
def findmissingpos(nums):
    n = len(nums)
    for i in range(n):
        while 1<=nums[i] <= n and nums[nums[i]-1] != nums[i]:
            # x 的位置放在 x-1 上
            nums[nums[i]-1], nums[i] = nums[i], nums[nums[i]-1]
    for i in range(n):
        if nums[i] != i+1:
            return i+1
    return n+1
```

leetcode 50 pow(x, n) 二进制 快速幂 log(n) O(1)
输入：x = 2.00000, n = 10 输出：1024.00000

```python
def mypow(x, n):
    def quickmul(N):
        ans = 1.0
        x_contri = x
        while N > 0:
            if N % 2 == 1:  # 如果 N 二进制表示的最低位为 1，那么需要计入贡献
                ans *= x_contri 
            x_contri *= x_contri
            N // 2 #舍弃最低位
        return ans
    return quickmul(n) if n >= 0 else 1.0/quickmul(-n)
```

leetcode 54 螺旋矩阵 O(m*n) O(m*n) 
给你一个 m 行 n 列的矩阵 matrix ，请按照 顺时针螺旋顺序 ，返回矩阵中的所有元素。

```python
def spiralOrder(matrix):
    if not matrix or not matrix[0]: return []
    m, n = len(matrix), len(matrix[0])
    res = []
    x, y = 0, 0
    left, right, up, down = 0, n-1, 0, m-1
    dirs = [(0,1),(1,0),(0,-1),(-1,0)] #右，下，左，上
    cur_d = 0
    while len(res) != m*n:
        res.append(matrix[x][y])
        if cur_d == 0 and y == right: #向右 而且 到了右边界
            cur_d += 1
            up += 1
        elif cur_d == 1 and x == down:
            cur_d += 1
            right -= 1
        elif cur_d == 2 and y == left:
            cur_d += 1
            down -= 1
        elif cur_d == 3 and x == up:
            cur_d += 1
            left += 1
        cur_d %= 4
        x += dirs[cur_d][0]
        y += dirs[cur_d][1]
    return res
```

leetcode 76 最小覆盖子串 O(n) O(1) 滑动窗口 + 哈希表
输入：s = "ADOBECODEBANC", t = "ABC" 输出："BANC"

```python
import collections
def minWindow(s, t):
    n, m = len(s), len(t)
    if n < m: return ""
    
    need = collections.defaultdict(int)
    for c in t:
        need[c] += 1
    needcnt = m # 用于记录当前剩余所需字母的数量
    res = [0, n] # 用于记录符合要求的滑窗的左右指针
    
    left = 0
    for right, c in enumerate(s):
        if need[c] > 0:
            needcnt -= 1
        need[c] -= 1 #如果出现某个字母的value为负数，则表示这个字母多余
        if needcnt == 0: #代表当前窗口包含了所需的所有字符, 收紧窗口 去除多余字符
            while True:
                if need[s[left]] == 0: # 该字符不能再减少了
                    break
                need[s[left]] += 1 #右移
                left += 1
            if right - left < res[1] - res[0]:
                res = [left, right] #当前字符串更小，更新res
    return s[res[0]:res[1]+1] if res != [0,n] else ""
```

leetcode 149 直线上最多的点数
对于每一个点，统计该点与其他点的斜率，用unordered_map的key存储斜率，用value存储个数
输入：points = [[1,1],[3,2],[5,3],[4,1],[2,3],[1,4]] 输出：4

```python
def maxpoints(points):
    n = len(points)
    if n < 3: return n
    res = 0
    for i in range(n):
        dic = {}
        same = 0
        curmax = 0
        for j in range(i+1, n):
            if points[i] == points[j]:
                same += 1
                continue
            if points[i][0] == points[j][0]:
                rate = float("inf")
            else:
                rate = (points[i][1]-points[j][1]) * 1000 / (points[i][0]-points[i][0]) * 1000
        dic[rate] = dic.get(rate, 0) + 1
        curmax = max(curmax, dic[rate])
    res = max(res, curmax+same+1)
    return res
```

```python
def searchMatrix(matrix, target):
    if len(matrix) == 0 or len(matrix[0]) == 0: return False
    n = len(matrix)
    m = len(matrix[0])
    left = 0
    right = m*n - 1
    while left <= right:
        mid = (left + right)//2
        if matrix[mid//m][mid%m] == target:
            return True
        if matrix[mid//m][mid%m] > target:
            right = mid - 1
        else:
            left = mid + 1
    return False
def searchMatrix(matrix, target):
    row = len(matrix)
    col = len(matrix[0])
    i = 0
    j = col - 1
    while matrix[i][j] != target: # 第一排最后一个开始
        if matrix[i][j] < target:
            i += 1
        else:
            j -= 1
        if i >= row or j < 0:
            return False
    return True
```

```python
def coinchange(coins, amount):
    dp = [float('inf')]*(amount+1) #dp[i] 组成金额i需要的最小硬币数量
    dp[0] = 0
    # F(2) = min(F(2-1),F(2-2),F(2-5)+1)
    for coin in coins:
        for x in range(coin, amount+1):
            dp[x] = min(dp[x], dp[x-coin]+1)
    return dp[amount] if dp[amount] != float('inf') else -1
```

```python
# leetcode 264 offer 49 丑数（只包含质因子2，3和5）求第n个丑数 O(N) O(N)
# 输入: n = 10 输出: 12 解释: 1, 2, 3, 4, 5, 6, 8, 9, 10, 12 是前 10 个丑数。
def nthuglynumber(n):
    if n < 0: return 0
    dp = [1] * n
    index2, index3, index5 = 0,0,0
    for i in range(1,n):
        dp[i] = min(dp[index2]*2, dp[index3]*3, dp[index5]*5)
        if dp[i] == 2*dp[index2]: index2 += 1
        if dp[i] == 3*dp[index3]: index3 += 1
        if dp[i] == 5*dp[index5]: index5 += 1
    return dp[n-1]
```

```python
# leetcode 1190 反转每对括号间的子串
# 输入：s = "(u(love)i)" 输出："iloveu"
def reverseP(s):
    stack = []
    for c in s:
        if c != ')':
            stack.append(c)
        elif c == ')':
            tmp = []
            while stack and stack[-1] != '(': #在遇到（之前 pop一个append一个 反转
                tmp.append(stack.pop())
            if stack:
                stack.pop()
            stack += tmp
    return "".join(stack)
```

```python
# leetcode 739 每日温度 O(N) O(N)
# 请计算在每一天需要等几天才会有更高的温度。如果气温在这之后都不会升高，用0代替 
# 输入: temperatures = [73,74,75,71,69,72,76,73] 输出: [1,1,4,2,1,1,0,0]
def dailyTemperatures(temperatures):
    ln = len(temperatures)
    ret = [0] * ln
    stack = []
    for i in range(ln):
        while stack and temperatures[i] > temperatures[stack[-1]]:
            index = stack.pop()
            ret[index] = i - index #记录index差值
        stack.append(i)
        #当栈为空或当前index对应的温度，小于栈顶index对应的温度时，index入栈
    return ret
```

```python
# leetcode 1047. 删除字符串中的所有相邻重复项
# 输入："abbaca" 输出："ca"
def removeDuplicates(self, s: str) -> str:
    if len(s) == 1: return s
    stack = []
    for c in s:
        if stack and c == stack[-1]:
            stack.pop()
        else:
            stack.append(c)
    return "".join(stack)
```

```python
# leetcode 329 矩阵中的最长递增路径
def longestIncreasingPath(matrix):
    if not matrix: return 0
    m, n = len(matrix), len(matrix[0])
    record = [[0] * n for _ in range(m)]
    res = 0
    def dfs(i, j):
        nonlocal res
        compare = []
        for dx, dy in [[1, 0], [-1, 0], [0, 1], [0, -1]]:
            x, y = i + dx, j + dy
            if 0 <= x < m and 0 <= y < n and matrix[x][y] > matrix[i][j]:
                if record[x][y]:
                    compare.append(record[x][y])
                else:
                    compare.append(dfs(x, y))
        record[i][j] = max(compare) + 1 if compare else 1
        res = max(res, record[i][j])
        return record[i][j]
    for i in range(m):
        for j in range(n):
            if not record[i][j]:
                dfs(i, j)
    return res
```

```python
# leetcode 334  递增的三元子序列
def increasingTriplet(nums):
    if len(nums) <= 2: return False
    small, mid = float('inf'), float('inf')
    for x in nums:
        if x <= small:
            small = x
        elif x <= mid:
            mid = x
        elif x > mid:
            return True 
    return False
```
