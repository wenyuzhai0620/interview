```python
#选择排序 n^2
def selection_sort(nums):
    for i in range(len(nums)):
        for j in range(i, len(nums)):
            if nums[i] > nums[j]:
                nums[i], nums[j] = nums[j], nums[i]
    return nums
```

```python
#冒泡排序 n^2 相邻元素之间的比较，将大的数慢慢“沉底”
def bubble_sort(nums):
    for i in range(len(nums)-1):
        for j in range(len(nums)-i-1):
            if nums[j] > nums[j+1]:
                nums[j], nums[j+1] = nums[j+1], nums[j]
    return nums

#优化 任意两数之间没有发生交换，如果任意两数之间有序，全部有序
def bubble_sort1(nums):
    for i in range(len(nums)-1):
        flag = 0
        for j in range(len(nums)-i-1):
            if nums[j] > nums[j+1]:
                nums[j], nums[j+1] = nums[j+1], nums[j]
                flag = 1
        if not flag: break
    return nums
```

```python
# 插入排序 n^2 从前面已排序数组找到插入的位置
def insertion_sort(nums):
    for i in range(1, len(nums)):
        while i > 0 and nums[i-1]>nums[i]:
            nums[i-1], nums[i] = nums[i], nums[i-1]
            i -= 1
    return nums
```

```python
# 希尔排序 n^1.3
def shell_sort(nums):
    n = len(nums)
    gap = n // 2
    while gap:
        for i in range(gap, n):
            while i - gap >= 0 and nums[i - gap] > nums[i]:
                nums[i - gap], nums[i] = nums[i], nums[i - gap]
                i -= gap
        gap //= 2
    return nums
```

```python
# 归并排序 nlogn
def merge_sort(nums):
    if len(nums) <= 1:
        return nums
    mid = len(nums) // 2
    left = merge_sort(nums[:mid])
    right = merge_sort(nums[mid:])
    return merge(left, right)

def merge(left, right):
    res = []
    i = 0
    j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            res.append(left[i])
            i += 1
        else:
            res.append(right[j])
            j += 1
    res += left[i:]
    res += right[j:]
    return res
```

```python
#快排 nlogn 最差：倒序 退化成冒泡
def quick_sort(nums):
    n = len(nums)
    def quick(left, right):
        if left >= right: return nums
        pivot = left
        i = left
        j = right
        while i < j:
            while i < j and nums[j] > nums[pivot]:
                j -= 1
            while i < j and nums[i] <= nums[pivot]:
                i += 1
            nums[i], nums[j] = nums[j], nums[i]
        nums[pivot], nums[j] = nums[j], nums[pivot]
        quick(left, j - 1)
        quick(j + 1, right)
        return nums
    return quick(0, n - 1)
```

```python
#offer 40 最小的k个数
def getLeastNumbers(arr, k):
    if k >= len(arr): return arr
    def quick_sort(l, r):
        i, j = l, r
        while i < j:
            while i < j and arr[j] >= arr[l]: j -= 1
            while i < j and arr[i] <= arr[l]: i += 1
            arr[i], arr[j] = arr[j], arr[i]
        arr[l], arr[i] = arr[i], arr[l]
        if k < i: return quick_sort(l, i - 1) 
        if k > i: return quick_sort(i + 1, r)
        return arr[:k]

    return quick_sort(0, len(arr) - 1)

def getLeastNumbers(arr, k):
    def build_heap(heap, i): #o(N)
        left, right = 2*i+1, 2*i+2
        large_index = i
        if left < len(heap) and heap[i] < heap[left]:
            large_index = left
        if right < len(heap) and heap[large_index] < heap[right]:
            large_index = right
        #通过比较上面和左右节点，得出三个元素中最大的
        if large_index != i:
            heap[i], heap[large_index] = heap[large_index], heap[i]
            build_heap(heap, large_index)
    
    if not arr or k <= 0: return []
    if len(arr) <= k: return arr
    heap = arr[:k]
    for i in range(len(heap) // 2 - 1, -1, -1):
        build_heap(heap, i) # 构建大根堆
    for i in range(k, len(arr)):
        if arr[i] < heap[0]:  # 当前元素比堆根小，则弹出堆根并插入当前元素
            heap[0] = arr[i]
            build_heap(heap, 0) # 重新调整堆根
    return heap
```

```python
# leetcode 347 前k个高频元素
#输入: nums = [1,1,1,2,2,3], k = 2 输出: [1,2]
import random
from collections import Counter
def topKFrequent(nums, k):
    lookup = Counter(nums)
    nums = [(freq, num) for num, freq in lookup.items()]
    # [(3, 1), (2, 2), (1, 3)]
    # 返回的是最大的前k个元组
    kths = findKth(nums, k, 0, len(nums)-1)
    print(kths)
    return [num for (freq, num) in kths]

# 找到第k大的数所在的位置，返回nums中k及其前面全部的元素；时间复杂度O(n)
def findKth(nums, k, low, high) :
    if low >= high: return nums[:low+1]
    pivot = random.randint(low, high)
    nums[low], nums[pivot] = nums[pivot], nums[low]
    base = nums[low][0]
    i = low
    for j in range(low+1, high+1) :
        if nums[j][0] > base :
            nums[j], nums[i+1] = nums[i+1], nums[j]
            i += 1
    nums[low], nums[i] = nums[i], nums[low]
    if i == k-1 :
        return nums[:i+1]
    elif i > k - 1 :
        return findKth(nums, k, low, i-1)
    else:
        return findKth(nums, k, i+1, high)
```

```python
# leetcode 378 有序矩阵中第 K 小的元素
def kthSmallest(matrix, k):
    n = len(matrix)
    def check(mid):
        i, j = n-1, 0
        num = 0
        while i >= 0 and j < n:
            if matrix[i][j] <= mid:
                num += i+1
                j += 1
            else:
                i -= 1
        return num >= k

    left, right = matrix[0][0], matrix[-1][-1]
    while left < right:
        mid = (left + right)//2
        if check(mid):
            right = mid
        else:
            left = mid + 1
    return left
```
