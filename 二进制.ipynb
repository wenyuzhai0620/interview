leetcode 190 颠倒二进制位
输入: 00000010100101000001111010011100
输出: 00111001011110000010100101000000
解释: 输入的二进制串表示无符号整数 43261596，因此返回 964176192

```python
def reverseBits(self, n: int) -> int:
    res = 0
    for i in range(32):
        res = (res << 1) | (n & 1) 
        #每次把 res 左移，把 n 的二进制末尾数字拼接到结果 res 的末尾
        n >>= 1 #把n 右移
    return res
```

leetcode 191 位1的个数
输入：0000001011 (输入是一个无符号整数（以二进制串的形式))
输出：3

```python
def hammingWeight(self, n: int) -> int:
    res = 0
    while n:
        res += 1
        n &= n-1 #n & 1取最后一位
    return res
```
