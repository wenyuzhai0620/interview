leetcode 116 填充每个节点的下一个右侧节点指针 完美二叉树填充


```python
def connect(self, root: 'Node') -> 'Node':
    if not root:
        return root
    leftmost = root
    while leftmost.left:  # 循环条件是当前节点的left不为空，当只有根节点
        head = leftmost
        while head:
            head.left.next = head.right
            if head.next:
                head.right.next = head.next.left
            head = head.next #继续向右遍历
        leftmost = leftmost.left #从下一层的左边开始遍历  
    return root
```

leetcode 124 二叉树的最大路径和 

```python
def maxPathSum(root):
    maxSum = float("-inf") 
    def maxgain(node):
        if not node: return 0
        leftgain = max(maxgain(node.left),0)
        rightgain = max(maxgain(node.right),0)

        pricenewpath = node.val + leftgain + rightgain

        maxSum = max(maxSum, pricenewpath)

        return node.val + max(leftgain, rightgain)

    maxgain(root)
    return maxSum
```

```python
# leetcode 230 二叉搜索树中第k小的元素 中序 O(n) O(n)
# offer 54  二叉搜索树的第k大节点
def kthSmallest(root, k):
    res = []
    def dfs(root, res):
        if not root:
            return
        dfs(root.left, res)
        res.append(root.val)
        dfs(root.right, res)
    dfs(root, res)
    return res[k-1] #  res[-k] 第k大
```

```python
def lowestCommonAncestor(root, p, q):
    ancestor = root
    while True:
        if p.val < ancestor.val and q.val < ancestor.val: 
            ancestor = ancestor.left
        elif p.val > ancestor.val and q.val > ancestor.val: 
            ancestor = ancestor.right
        else: break
    return ancestor
def lowestCommonAncestor(root, p, q):
    if not root or root == p or root == q: return root
    left = self.lowestCommonAncestor(root.left, p, q)
    right = self.lowestCommonAncestor(root.right, p, q)
    if not left: return right
    if not right: return left
    return root
```

```python
#leetcode 110 offer 55 II 平衡二叉树 O(n)  O(n)
def isBalanced(root):
    def dfs(node, depth):
        if not node: return 0
        left = dfs(node.left, depth+1)
        right = dfs(node.right, depth + 1)
        return max(left,right)+1

    if not root: return True
    if abs(dfs(root.left,0) - dfs(root.right,0)) > 1:return False
    return self.isBalanced(root.left) and self.isBalanced(root.right)
```

leetcode 863  二叉树中所有距离为 K 的结点
输入：root = [3,5,1,6,2,0,8,null,null,7,4], target = 5, K = 2
输出：[7,4,1]

```python
def distanceK(root, target, k):
    maps={}#构建哈希表，记录每个节点的父亲节点
    def findP(root):
        if root.left:
            maps[root.left.val]=root
            findP(root.left)
        if root.right:
            maps[root.right.val]=root
            findP(root.right)
    res=[]
    def dfs(root,node,k):
        #需要记录上一个遍历的节点是啥，防止5-->3 然后 3又去左子树5  这样走了两步
        if k==0:
            res.append(root.val)
            return 
        if root.left and root.left!=node:#左
            dfs(root.left,root,k-1)
        if root.right and root.right!=node:#右
            dfs(root.right,root,k-1)
        if root.val in maps and maps[root.val]!=node:#父亲
            dfs(maps[root.val],root,k-1)
    findP(root)
    dfs(target,None,k)
    return res
```

```python
# leetcode 199 二叉树的右视图 dfs
#　输入: [1,2,3,null,5,null,4] 输出: [1,3,4]
def rightSideView(self, root: TreeNode) -> List[int]:
    rightmost_value_at_depth = dict() # 深度为索引，存放节点的值
    max_depth = -1
    stack = [(root, 0)]
    while stack:
        node, depth = stack.pop()
        if node is not None:
            # 维护二叉树的最大深度
            max_depth = max(max_depth, depth)
            # 如果不存在对应深度的节点我们才插入
            rightmost_value_at_depth.setdefault(depth, node.val)
            stack.append((node.left, depth + 1))
            stack.append((node.right, depth + 1))
    return [rightmost_value_at_depth[depth] for depth in range(max_depth + 1)]
```
